```python
# 1. Basic Role Counts (from person's job titles) by scanning each person's job titles for these keywords and counted matches.
#     serial_founder_count: How many times the person has "founder" in their job titles
#     serial_advisor_count: How many times "advisor" appears in their job titles
#     serial_board_member_count: Count of "board" mentions in their job titles
#     serial_investor_count: How many times "investor" appears in their roles
#     serial_executive_count: Count of executive roles (CEO/CTO/CFO/etc.) in their titles

# 2. Organization Metrics by trackking their company relationships and job history over time
#     serial_num_founded_organizations: How many companies the person has founded
#     serial_num_current_jobs: Number of jobs they currently hold
#     serial_num_past_jobs: Number of past jobs in their history
#     serial_num_exits: How many companies they founded that got acquired/closed
#     serial_num_exits_ipo: How many companies they founded that went IPO
#     serial_num_past_company_acquisition: Companies they founded that were acquired

# 3. Gender-Based Counts by analyzing team composition at their companies by gender.
#     serial_num_male_current_jobs: Male colleagues in current workplaces
#     serial_num_female_current_jobs: Female colleagues in current workplaces
#     serial_num_male_founder: Male co-founders they've worked with
#     serial_num_female_founder: Female co-founders they've worked with
#     serial_num_male_board_member_jobs: Male board members they've served with
#     serial_num_female_board_member_jobs: Female board members they've served with
#     male_Executive_member_Count: Male executives they've worked with
#     female_Executive_member_Count: Female executives they've worked with

# 4. Location Metrics by checking where their colleagues were based.
#     executive_US_people: How many US-based executives they've worked with
#     executive_other_countries_people: Executives from other countries they've worked with
#     has_executive_other_countries: 1 if worked with international execs, else 0
#     has_executive_us_people: 1 if worked with US execs, else 0

# 5. Derived fields:
#     male_Founders_member_count: Same as serial_num_male_founder
#     female_Founders_member_count: Same as serial_num_female_founder
#     male_Board_mamber_count: Same as serial_num_male_board_member_jobs
#     female_Board_mamber_count: Same as serial_num_female_board_member_jobs
#     male_Current_Staffs: Same as serial_num_male_current_jobs
#     female_Current_Staffs: Same as serial_num_female_current_jobs
```

```python
import pandas as pd
import numpy as np

# Read all required Parquet files
df_people = pd.read_parquet(people_parquet)
df_organizations = pd.read_parquet(organizations_parquet)
df_jobs = pd.read_parquet(jobs_parquet)
df_funding_rounds = pd.read_parquet(funding_rounds_parquet)
df_acquisitions = pd.read_parquet(acquisitions_parquet)

# Fix boolean columns
df_jobs['is_current'] = df_jobs['is_current'].astype(bool)

# 1. Role counts from featured_job_title
df_people_roles = df_people.copy()
for role, pattern in [
    ('founder_count', r'\bfounder\b'),
    ('advisor_count', r'\badvisor\b'),
    ('board_member_count', r'\bboard\b'),
    ('investor_count', r'\binvestor\b'),
    ('executive_count', r'\bexecutive\b|ceo|cto|cfo|coo|chief')
]:
    df_people_roles[f"{role}"] = np.where(
        df_people_roles['featured_job_title'].str.lower().str.contains(pattern, na=False),
        1,
        0
    )

# 2. Founder relationships
df_founder_relationships = df_jobs[
    df_jobs['job_type'].str.lower().str.contains('founder', na=False) |
    df_jobs['title'].str.lower().str.contains('founder', na=False)
][['person_uuid', 'org_uuid']].rename(columns={
    'person_uuid': 'person_identifier',
    'org_uuid': 'organization_identifier'
})

# 3. Gender metrics
df_people_gender = df_people[['uuid', 'gender']]

# 4. Job information
df_jobs_filtered = df_jobs[['person_uuid', 'is_current', 'job_type', 'org_uuid']]
df_jobs_with_gender = pd.merge(
    df_jobs_filtered,
    df_people_gender,
    left_on='person_uuid',
    right_on='uuid',
    how='left'
)

# Helper function to process gender metrics - FIXED SYNTAX
def prepare_gender_metrics(df, group_col, prefix):
    """Process gender metrics with all categories and standardize column names"""
    gender_counts = (
        df.groupby(group_col)['gender']
        .value_counts()
        .unstack(fill_value=0)
    )

    # Standard column naming for all possible gender categories
    col_mapping = {
        'male': f'{prefix}_male',
        'female': f'{prefix}_female',
        'transgender_person': f'{prefix}_transgender',
        'androgynous': f'{prefix}_androgynous',
        'agender': f'{prefix}_agender',
        'not_provided': f'{prefix}_not_provided',
        'two_spirit': f'{prefix}_two_spirit',
        'other': f'{prefix}_other',
        'bigender': f'{prefix}_bigender',
        'androgyne': f'{prefix}_androgyne',
        'non_binary': f'{prefix}_non_binary',
        'mtf': f'{prefix}_mtf'
    }

    # Only keep columns that exist in the data
    existing_cols = {k: v for k, v in col_mapping.items() if k in gender_counts.columns}
    return gender_counts.rename(columns=existing_cols)

# 5. Calculate all metrics
metrics = {
    # Current jobs by gender
    'current_jobs': prepare_gender_metrics(
        df_jobs_with_gender[df_jobs_with_gender['is_current']],
        'person_uuid', 'num_current_jobs'
    ),

    # Founder counts by gender
    'founders': prepare_gender_metrics(
        pd.merge(
            df_founder_relationships,
            df_people_gender,
            left_on='person_identifier',
            right_on='uuid',
            how='left'
        ),
        'person_identifier', 'num_founder'
    ),

    # Board members by gender
    'board_members': prepare_gender_metrics(
        df_jobs_with_gender[
            df_jobs_with_gender['job_type'].str.lower().str.contains('board|director', na=False)
        ],
        'person_uuid', 'num_board_member_jobs'
    ),

    # Executive counts by gender
    'executives': prepare_gender_metrics(
        df_jobs_with_gender[
            df_jobs_with_gender['job_type'].str.lower().str.contains('executive|ceo|cto|cfo|coo|chief', na=False)
        ],
        'person_uuid', 'num_executive'
    ),

    # Founded organizations count
    'founded_orgs': df_founder_relationships
        .groupby('person_identifier')['organization_identifier']
        .nunique()
        .rename('num_founded_organizations'),

    # Job counts
    'job_counts': df_jobs_filtered.groupby('person_uuid').agg(
        num_current_jobs=('is_current', 'sum'),
        num_past_jobs=('is_current', lambda x: (~x).sum()),
        primary_job_type=('job_type', 'max')
    ),

    # Exit metrics
    'exits': pd.merge(
        df_founder_relationships,
        df_organizations[['uuid', 'closed_on', 'status']].rename(columns={'uuid': 'acquired_by'}),
        left_on='organization_identifier',
        right_on='acquired_by',
        how='left'
    )
    .assign(
        has_exit=lambda x: x['closed_on'].notna(),
        is_ipo=lambda x: x['status'] == 'ipo',
        is_acquired=lambda x: x['acquired_by'].notna()
    )
    .groupby('person_identifier')
    .agg(
        num_exits=('has_exit', 'sum'),
        num_exits_ipo=('is_ipo', 'sum'),
        num_past_company_acquisition=('is_acquired', 'sum')
    ),

    # Location metrics
    'location': df_people[['uuid', 'country_code']]
        .assign(
            is_us=lambda x: x['country_code'] == 'USA',
            is_not_us=lambda x: x['country_code'] != 'USA'
        )
        .groupby('uuid')
        .agg(
            executive_us=('is_us', 'sum'),
            executive_other=('is_not_us', 'sum')
        )
}

# 6. Combine all metrics with proper suffixes
df_metrics = (
    df_people[['uuid', 'name', 'featured_job_organization_uuid']]
    .merge(df_people_roles.groupby(['uuid', 'name'])[
        ['founder_count', 'advisor_count', 'board_member_count', 'investor_count', 'executive_count']
    ].sum().reset_index(), on=['uuid', 'name'], how='left')

    # Merge metrics with suffixes to avoid duplicate columns
    .merge(metrics['current_jobs'], left_on='uuid', right_index=True, how='left', suffixes=('', '_current'))
    .merge(metrics['founders'], left_on='uuid', right_index=True, how='left', suffixes=('', '_founder'))
    .merge(metrics['board_members'], left_on='uuid', right_index=True, how='left', suffixes=('', '_board'))
    .merge(metrics['executives'], left_on='uuid', right_index=True, how='left', suffixes=('', '_exec'))
    .merge(metrics['founded_orgs'], left_on='uuid', right_index=True, how='left')
    .merge(metrics['job_counts'], left_on='uuid', right_index=True, how='left')
    .merge(metrics['exits'], left_on='uuid', right_index=True, how='left')
    .merge(metrics['location'], left_on='uuid', right_index=True, how='left')
)

# Define column renaming mapping
column_rename_map = {
    'founder_count': 'serial_founder_count',
    'advisor_count': 'serial_advisor_count',
    'board_member_count': 'serial_board_member_count',
    'investor_count': 'serial_investor_count',
    'executive_count': 'serial_executive_count',
    'num_founded_organizations': 'serial_num_founded_organizations',
    'num_current_jobs': 'serial_num_current_jobs',
    'num_past_jobs': 'serial_num_past_jobs',
    'num_exits': 'serial_num_exits',
    'num_exits_ipo': 'serial_num_exits_ipo',
    'num_past_company_acquisition': 'serial_num_past_company_acquisition',
    'num_current_jobs_male': 'serial_num_male_current_jobs',
    'num_current_jobs_female': 'serial_num_female_current_jobs',
    'num_founder_male': 'serial_num_male_founder',
    'num_founder_female': 'serial_num_female_founder',
    'num_board_member_jobs_male': 'serial_num_male_board_member_jobs',
    'num_board_member_jobs_female': 'serial_num_female_board_member_jobs',
    'num_executive_male': 'male_Executive_member_Count',
    'num_executive_female': 'female_Executive_member_Count',
    'executive_us': 'executive_US_people',
    'executive_other': 'executive_other_countries_people'
}

# Apply renaming and keep only needed columns
df_metrics = df_metrics.rename(columns=column_rename_map)

# Select only the columns we want in the final output
final_columns = [
    'uuid', 'name', 'featured_job_organization_uuid',
    'serial_founder_count', 'serial_advisor_count', 'serial_board_member_count',
    'serial_investor_count', 'serial_executive_count', 'serial_num_founded_organizations',
    'serial_num_current_jobs', 'serial_num_past_jobs', 'primary_job_type',
    'serial_num_exits', 'serial_num_exits_ipo', 'serial_num_past_company_acquisition',
    'serial_num_male_current_jobs', 'serial_num_female_current_jobs',
    'serial_num_male_founder', 'serial_num_female_founder',
    'serial_num_male_board_member_jobs', 'serial_num_female_board_member_jobs',
    'male_Executive_member_Count', 'female_Executive_member_Count',
    'executive_US_people', 'executive_other_countries_people'
]

df_metrics = df_metrics[final_columns]

# Fill NA values appropriately
for col in df_metrics.columns:
    if col in ['uuid', 'name', 'featured_job_organization_uuid', 'primary_job_type']:
        df_metrics[col] = df_metrics[col].fillna('')
    else:
        df_metrics[col] = df_metrics[col].fillna(0).astype(int)

# Add derived columns
df_metrics = df_metrics.assign(
    has_executive_other_countries=lambda x: (x['executive_other_countries_people'] > 0).astype(int),
    has_executive_us_people=lambda x: (x['executive_US_people'] > 0).astype(int),
    male_Founders_member_count=lambda x: x['serial_num_male_founder'],
    female_Founders_member_count=lambda x: x['serial_num_female_founder'],
    male_Board_mamber_count=lambda x: x['serial_num_male_board_member_jobs'],
    female_Board_mamber_count=lambda x: x['serial_num_female_board_member_jobs'],
    male_Current_Staffs=lambda x: x['serial_num_male_current_jobs'],
    female_Current_Staffs=lambda x: x['serial_num_female_current_jobs']
)

# Show final dataframe
print(df_metrics.head())

df_metrics.to_csv("")
```
